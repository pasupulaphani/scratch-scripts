```python
import pandas as pd
import numpy as np
import yfinance as yf
from ta import add_all_ta_features
from datetime import datetime, timedelta
from fredapi import Fred
import calendar

# === CONFIG ===
FRED_API_KEY = "a5227ed5308740308ac048c76c3659b6"
fred = Fred(api_key=FRED_API_KEY)


transcripts_df = pd.read_csv('data/earnings_call_train.csv')

transcripts_df["date"] = pd.to_datetime(transcripts_df["date"])


TICKERS = transcripts_df["company"].dropna().unique().tolist()
START_DATE = transcripts_df["date"].min() - pd.Timedelta(days=365)
END_DATE = transcripts_df["date"].max() + pd.Timedelta(days=30)

# === MACRO DATA (FRED) ===
def get_macro_data():
    dates = pd.date_range(START_DATE, END_DATE)
    df = pd.DataFrame(index=dates)
    df["ffr"] = fred.get_series("FEDFUNDS", observation_start=START_DATE).reindex(df.index).ffill()
    df["cpi"] = fred.get_series("CPIAUCSL", observation_start=START_DATE).reindex(df.index).ffill()
    df["unemployment"] = fred.get_series("UNRATE", observation_start=START_DATE).reindex(df.index).ffill()
    df["vix"] = fred.get_series("VIXCLS", observation_start=START_DATE).reindex(df.index).ffill()
    return df.reset_index().rename(columns={"index": "date"})

macro_df = get_macro_data()

# === COMPANY FEATURES ===
all_data = []

for ticker in TICKERS:
    print(f"Processing {ticker}")
    df = yf.download(ticker, start=START_DATE, end=END_DATE, multi_level_index=False)
    df = df.reset_index().rename(columns={"Date": "date"})
    df["ticker"] = ticker

    # === Technical Indicators ===
    df = add_all_ta_features(
        df,
        open="Open", high="High", low="Low", close="Close", volume="Volume",
        fillna=True
    )

    # === Price/Volume Features ===
    df["return_1d"] = df["Close"].pct_change()
    df["return_5d"] = df["Close"].pct_change(5)
    df["return_10d"] = df["Close"].pct_change(10)
    df["volatility_5d"] = df["return_1d"].rolling(5).std()
    df["volatility_10d"] = df["return_1d"].rolling(10).std()
    df["gap_open_close"] = df["Open"] - df["Close"].shift(1)
    df["gap_high_low"] = df["High"] - df["Low"]
    df["momentum_5d"] = df["Close"] - df["Close"].shift(5)
    df["volume_change"] = df["Volume"].pct_change()
    df["volume_vs_avg"] = df["Volume"] / df["Volume"].rolling(10).mean()

    # === Calendar & Event Features ===
    df["day_of_week"] = df["date"].dt.dayofweek
    df["month"] = df["date"].dt.month
    df["quarter"] = df["date"].dt.quarter
    df["is_month_start"] = df["date"].dt.is_month_start.astype(int)
    df["is_month_end"] = df["date"].dt.is_month_end.astype(int)

    # === Merge Macro Data ===
    df = pd.merge_asof(df.sort_values("date"), macro_df.sort_values("date"), on="date")

    # === Placeholder: Fundamentals (Quarterly) ===
    fin = yf.Ticker(ticker)
    try:
        fundamentals = fin.quarterly_financials.T
        fundamentals["date"] = fundamentals.index
        df = pd.merge_asof(df.sort_values("date"), fundamentals.reset_index(drop=True), on="date")
    except:
        print(f" Fundamentals not available for {ticker}")

    # === Placeholder: News & Social Sentiment ===
    df["news_sentiment"] = np.nan
    df["social_sentiment"] = np.nan
    df["news_volume"] = np.nan
    df["social_volume"] = np.nan

    all_data.append(df)

# === Combine All ===
final_df = pd.concat(all_data).reset_index(drop=True)
print(final_df.columns)
final_df.to_csv("data/full_stock_features.csv", index=False)
print("All features save to data/full_stock_features.csv")
```
